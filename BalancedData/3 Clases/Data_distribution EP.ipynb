```python
# Load Dataset
PATH = '/content/Epileptic Seizure Recognition.csv'
#readdata = pd.read_csv(PATH)
#dataset = pd.read_csv(PATH, names=['Vsl', 'Vsg', 'VisL', 'VisG', 'DenL', 'DenG', 'ST', 'Ang', 'ID', 'Flow Pattern'], header=0)
dataset = pd.read_csv(PATH)
# Summarize the Dataset
print("shape of data =",dataset.shape)
# Class Distribution
print(dataset.groupby('y').size())
```

```python
fig, ax = plt.subplots()
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=12
cantidades = [419,  #B
              184,  #SS
              66,  #DB
              37,  #SW
              ]

nombres = ["ScienceDirect",
           "Springer",
           "Scopus",
           "IEEExplorer",
           ]

color_palette_list = ['#C1F0F6','#007ACD',"#FFD97D","#60D394","#EE6055"]

ax.pie(cantidades, labels=nombres,
       colors=color_palette_list[0:], autopct='%1.0f%%',
       shadow=False, startangle=0,
       pctdistance=1.2,labeldistance=1.4)
centre_circle = plt.Circle((0, 0), 0.4, color='white')
ax.add_artist(centre_circle)
ax.axis('equal')
ax.legend(frameon=False, bbox_to_anchor=(1.3,0.8))
ax.figure.subplots_adjust(right=0.8)
ax.figure.savefig(figures_path+"/MultiPart.pdf")
```
