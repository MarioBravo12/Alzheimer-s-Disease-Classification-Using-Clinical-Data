```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ["AD",
           "CN",
           "EMCI",
           "LMCI"]
valores = [1591,  #B
              4201,  #SS
              2527,  #DB
              4863  #SW

              ]

# Cálculo de los porcentajes
total = sum(valores)
porcentajes = [(valor / total) * 100 for valor in valores]

# Colores para cada sección del donut
colores = ['#C1F0F6','#007ACD',"#FFD97D","#60D394"]

# Crea un gráfico de donut
fig, ax = plt.subplots()
ax.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90, colors=colores, pctdistance=0.85)

# Dibuja un círculo en el centro para crear el agujero en el donut
circle = plt.Circle((0, 0), 0.7, color='white')
ax.add_artist(circle)

# Ajusta el aspecto y muestra el gráfico
ax.axis('equal')  # Para que el gráfico sea un círculo
plt.title("Gráfico de Donut con Valores y Porcentajes")

plt.show()
```

```python
fig, ax = plt.subplots()
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=12
cantidades = [1591,  #B
              4201,  #SS
              2527,  #DB
              4863,  #SW

              ]

nombres = ["AD",
           "CN",
           "EMCI",
           "LMCI",

           ]

total = sum(cantidades)
porcentajes = [(valor / total) * 100 for valor in cantidades]

color_palette_list = ['#C1F0F6','#007ACD',"#FFD97D","#60D394"]

ax.pie(cantidades, labels=nombres,
       colors=color_palette_list[0:], autopct='%1.0f%%',
       shadow=False, startangle=0,
       pctdistance=1.2,labeldistance=1.4)
centre_circle = plt.Circle((0, 0), 0.4, color='white')
ax.add_artist(centre_circle)
ax.axis('equal')
ax.legend(frameon=False, bbox_to_anchor=(1.3,0.8))
ax.figure.subplots_adjust(right=0.8)
ax.figure.savefig("/MultiPart.pdf")
```
